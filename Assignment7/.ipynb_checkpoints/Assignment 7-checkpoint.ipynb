**Let $\bar{x} = (x_1, x_2, x_3, \cdots, x_m)$ denote a data point with $m$ features. Suppose we approximate $\bar{x}$ with fewer features, say two features $x_1$ and $x_2$, as $\bar{y} = (x_1, x_2, 0, \cdots, 0)$. Now, define the loss due to the representation as $\lVert \bar{x} - \bar{y} \rVert^2 = x_3^2 + \cdots + x_m^2$. **

**(1b) Find a single feature representation of the data (in data1.csv) such that the loss in representation is minimum.**

```python
# Write Code here
def LossInRep(x_original, x_transformed):
    loss = np.linalg.norm(x_original - x_transformed)**2
    return loss

X = np.array(df)
M = X.shape[0]
# Using the first feature
x_transformed1 = np.c_[X[:, 0], np.zeros(M)]
loss1 = LossInRep(X, x_transformed1)
print("Loss in representation using feature 1 = {}".format(loss1))

# Using the second feature
x_transformed2 = np.c_[np.zeros(M), X[:, 1]]
loss2 = LossInRep(X, x_transformed2)
print("Loss in representation using feature 2 = {}".format(loss2))
```

The loss in representation is minimum when only feature 1 is used. Therefore a good representation using only one feature is to use feature 1.

**(2b) Plot the minimum representation loss as a function of the number of features used for representation. Vary the number of features from 0 to 3**

I will consider all possible combinations of features and pick the best combination( the one that gives minimum loss in representation) for each "number of features".

```python
def CheckMinLossCombos(X, features, num_to_select):
    combos = list(combinations(features, num_to_select))
    min_loss = np.inf
    for combo in combos:
        x_transformed = np.zeros(X.shape)
        for feature in combo:
            x_transformed[:, feature] = X[:, feature]
        current_loss = LossInRep(X, x_transformed)
        if current_loss < min_loss:
            min_loss = current_loss
    return min_loss

features = [0, 1, 2, 3]
num_to_select = [0, 1, 2, 3]
loss_values = []
X = np.array(df2)
for num in num_to_select:
    loss_values.append(CheckMinLossCombos(X, features, num))
    print("Loss in representation when we use {} features = {}".format(num, loss_values[-1]))
    
plt.plot(features, loss_values, 'bo-')
plt.title("Loss in Representation vs Number of features", size=16)
plt.xlabel("Number of features", size=14)
plt.ylabel("Loss in Representation", size=14)
plt.xticks([0, 1, 2, 3])
plt.show()
```

As the number of features used increases, the loss in representation decreases.

**(2c) Order the features in decreasing order of significance (with respect to the representation loss).**

```python
features = [0, 1, 2, 3]
rep_loss = [] # Stores the representation loss when a certain feature is not used.
for feature in features:
    X = np.array(df2)
    x_transformed = np.array(df2)
    x_transformed[:, feature] = np.zeros(X.shape[0])
    rep_loss.append(LossInRep(X, x_transformed))
    
rep_loss_temp = [(rep_loss[i], i) for i in range(len(features))]
order_of_imp = sorted(rep_loss_temp, reverse=True)
print("Loss in representation without feature:")
print(*order_of_imp, sep="\n")
indices = [t[1] for t in order_of_imp]
print("Decreasing order of importance: {}".format(indices))
```

The loss value being printed shows the loss in representation incurred due to not including a particular column/feature. Thus a large loss in representation indicates that the column we neglected is very important.

**(2d) Compute the covariance matrix of the data. Give your observations between the values in covariance matrix and feature importance obtained in (2c).**

```python
# Write code here
X = np.array(df2)
xmeans = np.mean(X, axis=0).reshape(1, -1)
Xs = X - xmeans
M = X.shape[0]
cov_mat = (1/M) * np.dot(Xs.T, Xs)
print("Covariance Matrix:")
print(cov_mat)
```

The diagonal elements represent the covariance of the feature with itself, i.e. the variance of the feature. The descending order of the variances for features is: var(feature 2) > var(feature 1) > var(feature 0) > var(feature 3). This order is the same as the descending order of feature importance.

It shows that the variance gives us information about the variability of the data in that particular dimension. A high variance indicates a large variability, meaning that the feature is very important. Small variance indicates that there is not much variability in that dimension, so the loss in representation even after removing it will be low.

**(3c) Apply PCA to transform data by retaining only 1 PC. Now, visualize the data points in transformed 1-dimensional space**

```python
# Write code here
eigen_matrix = np.linalg.eig(cov_mat)
eigenvalues = eigen_matrix[0]
eigenvectors = eigen_matrix[1]

def PickMEigenVectors(eigenvalues, eigenvectors, m):
    n = len(eigenvalues)
    temp = [(eigenvalues[i], eigenvectors[:, i]) for i in range(n)]
    eigen = sorted(temp, reverse=True)
    # best m eigenvalues
    eigen_values_best = [t[0] for t in eigen[:m]]
    
    # best m eigenvectors
    eigen_vectors_best = [t[1] for t in eigen[:m]]
    
    # return only the first m (eigenvectors corresponding to largest m eigenvalues)
    return eigen_values_best, eigen_vectors_best

m = 1
eigen_values_best, eigen_vectors_best = PickMEigenVectors(eigenvalues, eigenvectors, m)

# Store the eigenvectors corresponding to "m" largest eigenvalues in V
V = np.c_[eigen_vectors_best[0]]
for vec in eigen_vectors_best[1:]:
    V = np.c_[V, vec]
    
# Data projected to 1 dimensional space
projection = np.dot(X, V)
```

**(3d) Compute the representational loss (with respect to the original data).**

```python
# On using all the principal components
m = 2
eigen_values_best, eigen_vectors_best = PickMEigenVectors(eigenvalues, eigenvectors, m)

# Store all eigenvectors in descending order of eigenvalue in V
V = np.c_[eigen_vectors_best[0]]
for vec in eigen_vectors_best[1:]:
    V = np.c_[V, vec]
    
# Data projected to 2 dimensional space
projection_all = np.dot(X, V)
    
# Data projected to 1-dimensional space
projection1 = np.dot(X, V[:, 0].reshape(-1, 1))

# Loss in representation
def LossInRep(x_original, x_transformed):
    loss = np.linalg.norm(x_original - x_transformed)**2
    return loss

M = X.shape[0]
# Using only the first feature
projection1_trans = np.c_[projection1, np.zeros(M)]
loss = LossInRep(projection_all, projection1_trans)
print("Loss in representation using only one principal component = {}".format(loss))
```

### Application of PCA on real dataset

**(4a) Load the human face data set from sklearn (fetch_lfw_people).**

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
```

**(4b) Apply PCA on the faces data (sklearn can be used). Take the top 150 features from PCA. Reconstruct the images with the these 150 features and visualise them. Give your observations.**

```python
# To select one example from each of the 8 classes
indices = []
number_of_targets = faces['target_names'].shape[0]
for t in range(number_of_targets):
    indices.append(list(faces['target']).index(t))
```

```python
# Write your code here
from sklearn.decomposition import PCA
X = np.array(faces['data'])

# Choosing 150 principal components
pca = PCA(n_components=150)
X_transformed = pca.fit_transform(X)
# Reconstructing the images
X_reconstruct = pca.inverse_transform(X_transformed).reshape(faces['images'].shape)
```

```python
f, axs = plt.subplots(8,2,figsize=(20, 15))
plt.subplots_adjust(left=None, bottom=None, right=0.4, top=1.5, wspace=0, hspace=None)
for i, ind in enumerate(indices):
    plt.subplot(8, 2, (2*i)+1)
    # Display the original image
    plt.imshow(X[ind].reshape(faces['images'][0].shape), cmap="gray")
    plt.title("Original: {}".format(faces['target_names'][i]))
    plt.xticks([])
    plt.yticks([])
    plt.subplot(8, 2, (2*i)+2)
    # Display the image after selecting 150 principal components
    plt.imshow(X_reconstruct[ind], cmap="gray")
    plt.title("After PCA")
    plt.xticks([])
    plt.yticks([])
```

### 5) Cross validation to select parameters for Ridge Regression 

$$\hat{\theta} = argmin_{\theta}((y - X\theta)^\top(y - X\theta) + \lambda \theta^\top \theta)$$


Perform ridge rigression using gradient descent on the data 4 provided. (You can use sklearn)

Try to fit the data with polynomial of degree 3 and 

- For $\lambda$ choose from = [10,100,1000]
- choose learning rate (here it is $\alpha$)= [ 0.01,0.001,0.0001]
- max_iters = 1000

Choose the best combination of $\lambda$ and $\alpha$ using 5-fold cross validation on the dataset provided and report the same.


```python
# Write your code here
from sklearn.linear_model import Ridge, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")

lambda_values = [10,100,1000]
lr_values = [0.01,0.001,0.0001]
max_iters = 1000 # Maximum number of iterations

# Building the parameter dictionary
parameter_grid = {"alpha": lambda_values, "eta0": lr_values}
```

```python
# Loading data into X_train and y_train
data = np.load("data4.npz")
X_train = data['X_train']
y_train = data['y_train']
```

```python
# To get a polynomial of degree 3
poly = PolynomialFeatures(degree=3)
X_transformed = poly.fit_transform(X_train.reshape(-1, 1))

# Gradient Descent model (using least square error and L2 regularization)
model = SGDRegressor(loss='squared_loss', penalty='l2', max_iter=max_iters)

# GridSearch 5-fold cross validation to find the best combination of parameters
grid_search = GridSearchCV(model, parameter_grid, cv=5)
grid_search.fit(X_transformed, y_train)
```

```python
best_lr = grid_search.best_params_['eta0']
best_lambda = grid_search.best_params_['alpha']
print("Best lambda value for L2 Regularization = {}".format(best_lambda))
print("Best learning rate for Gradient Descent = {}".format(best_lr))
```

```python
# Mean squared error for best combination of alpha and lambda
best_model = SGDRegressor(loss='squared_loss', penalty='l2', max_iter=max_iters, alpha=best_lambda, eta0=best_lr)
best_model.fit(X_transformed, y_train)
y_train_pred = best_model.predict(X_transformed)

# Calculate mean squared error for the best model(best combination of alpha and lambda)
train_loss = mean_squared_error(y_train, y_train_pred)
print("Mean Squared Error = {}".format(train_loss))
```

```python
x_values = np.linspace(min(X_train)-1, max(X_train)+1, 100).reshape(-1, 1)
x_values_transformed = poly.fit_transform(x_values)
y_values = best_model.predict(x_values_transformed)
plt.scatter(X_train, y_train, color='red', label='Training data')
plt.plot(x_values, y_values, color='blue', label='Learned function')
plt.title("Learned function and data points", size=16)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.legend()
plt.show()
```
